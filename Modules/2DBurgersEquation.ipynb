```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd

from input_variables import SetupParameters
from grid import MeshGrid
from difference_equations import DiscretizationSchemes
from matrix_init import MatrixInitializer
from physics import *
from boundary_conditions import BCs
from Visualization import *
from test import *

def main():
    params = SetupParameters(nx=41, ny=41, nt=120, c=1, sigma = 0.0009, nu=0.01)
    
    print("\nCustom Parameters:")
    print(f"Grid Size: {params.nx} x {params.ny}")
    print(f"Domain Size: {params.lx} x {params.ly}")
    print(f"Element Size: {params.dx} x {params.dy}")
    print(f"Number of iterations: {params.nt}")
    print(f"Time Step length: {params.dt}")
    print(f"Kinematic viscosity: {params.nu}")
    print(f"Fluid Density: {params.rho_fluid}")
    print(f"Reynolds number:{params.reynolds_number}")
    print(f"Prandtl Number: {params.prandtl_number}")
    print(f"Thermal Expansion of water: {params.beta}")
    print(f"alpha: {params.alpha}")
    print(f"gamma: {params.gamma}")
    print(f"sigma: {params.sigma}")
    
    X, Y, dx, dy =MeshGrid.mesh(params.nx, params.ny, params.lx, params.ly)
    MeshGrid.mesh_plot(X,Y)
    print("\nMesh Grid Information:")
    print(f"X shape: {X.shape}")
    print(f"Y shape: {Y.shape}")
    print(f"dx: {dx}")
    print(f"dy: {dy}")
    print(f"X range: {X.min()} to {X.max()}")
    print(f"Y range: {Y.min()} to {Y.max()}")
    
    ux_prev = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    ux_tent = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    ux_next = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    
    uy_prev = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    uy_tent = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    uy_next = MatrixInitializer.matrix_initialization(params.nx, params.ny, zero_initialization=False, initial_value=1)
    
    # Set hat function I.C.: u(.5<=x<=1 && .5<=y<=1) is 2
    ux_prev[int(.5 / params.dy):int(1 / params.dy + 1), int(.5 / params.dx):int(1 / params.dx + 1)] = 2
    uy_prev[int(.5 / params.dy):int(1 / params.dy + 1), int(.5 / params.dx):int(1 / params.dx + 1)] = 2
    
    # # this seciton is to extract the results
    # df_ux = pd.DataFrame(ux_prev)
    # df_uy = pd.DataFrame(uy_prev)

    # # Create an Excel writer object
    # with pd.ExcelWriter('C:\\Users\\saeed\\Desktop\\output_burger_initial.xlsx') as writer:
    #     # Write each DataFrame to a different sheet
    #     df_ux.to_excel(writer, sheet_name='ux_matrix', index=False, header=False)
    #     df_uy.to_excel(writer, sheet_name='uy_matrix', index=False, header=False)

    # print("Arrays exported to output.xlsx") 
    
    fig = plt.figure(figsize=(11,7), dpi=100)
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(X, Y, ux_prev[:], cmap=cm.viridis, rstride=1, cstride=1)
    ax.plot_surface(X, Y, uy_prev[:], cmap=cm.viridis, rstride=1, cstride=1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    
    Spin_up = []
    Error = []

    ux_solution_time = []
    uy_solution_time = []
    
    ux_solution_time_last = []
    uy_solution_time_last = []
    
    p_solution_time = []
    T_solution_time = []
    
    for _ in tqdm(range(params.nt)):
        # Discretization 
        # First equation
        dux_prev__dx = DiscretizationSchemes.backward_difference_x(ux_prev, params.dx)
        dux_prev__dy = DiscretizationSchemes.backward_difference_y(ux_prev, params.dy)
        laplace_ux_prev = DiscretizationSchemes.laplacian(ux_prev, params.dx, params.dy)
    
        # Second equation
        duy_prev__dx = DiscretizationSchemes.backward_difference_x(uy_prev, params.dx)
        duy_prev__dy = DiscretizationSchemes.backward_difference_y(uy_prev, params.dy)
        laplace_uy_prev = DiscretizationSchemes.laplacian(uy_prev, params.dx, params.dy)
        
        # print (duy_prev__dx)
        
        # Physics
        ux_tent = HomogeneousAdvection.advection_velocity_prediction_horizontal(
            ux_prev, uy_prev, dux_prev__dx, dux_prev__dy, laplace_ux_prev, 
            params.nu, params.dt
            )
        uy_tent = HomogeneousAdvection.advection_velocity_prediction_vertical(
            ux_prev, uy_prev, duy_prev__dx, duy_prev__dy, laplace_uy_prev, 
            params.nu, params.dt
         )
        
        ux_tent[0, :] = 1
        ux_tent[:, 0] = 1
        ux_tent[:, -1] = 1
        ux_tent[-1, :] = 1
        
        uy_tent[0, :] = 1
        uy_tent[:, 0] = 1
        uy_tent[:, -1] = 1
        uy_tent[-1, :] = 1
        
        ux_next = ux_tent
        uy_next = uy_tent
    
        # Advance in time
        ux_prev = ux_next
        uy_prev = uy_next
        
        ux_solution_time.append(ux_next)
        uy_solution_time.append(uy_next)
        
        # this seciton is to extract the results
        df_ux_next = pd.DataFrame(ux_next)
        df_uy_next = pd.DataFrame(uy_next)

        # Create an Excel writer object
        with pd.ExcelWriter('C:\\Users\\saeed\\Desktop\\output_burger_final4.xlsx') as writer:
            # Write each DataFrame to a different sheet
            df_ux_next.to_excel(writer, sheet_name='ux_next_matrix', index=False, header=False)
            df_uy_next.to_excel(writer, sheet_name='uy_next_matrix', index=False, header=False)

        print("Arrays exported to output.xlsx") 
        
        # Store only the last time step
        if _ == params.nt - 1:
            ux_solution_time_last = ux_next
            uy_solution_time_last = uy_next
        
# Plot only the last time step outside the loop
    fig = plt.figure(figsize=(11, 7), dpi=100)
    ax = fig.add_subplot(projection='3d')
    X, Y, dx, dy =MeshGrid.mesh(params.nx, params.ny, params.lx, params.ly)
    ax.plot_surface(X, Y, ux_next, cmap=cm.viridis, rstride=1, cstride=1)
    ax.plot_surface(X, Y, uy_next, cmap=cm.viridis, rstride=1, cstride=1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    plt.title(f'Results at final time step (t = {params.nt * params.dt})')
    plt.show()
            
        
if __name__ == "__main__":
    main()
    
```
